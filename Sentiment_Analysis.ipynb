```python
import snscrape.modules.twitter as sntwitter
```

```python
k='(#covid-19 OR #booster OR #dose) min_retweets:5000 lang:en until:2022-11-30 since:2020-03-20'
l=[]
for tweet in sntwitter.TwitterSearchScraper(k).get_items():
  if len(l)==100:
    break
  else:
    l.append(tweet.renderedContent)
```

```python
print(l)
```

## **Data** **Preprocessing** 


1 Cleaning(Removing symbols and punctuations)

[ #Character block start.
^ #not these characters(letters, numbers).
\w #word characters.
\s #space characters.
] #character block end.

```python
punctuations = '''!()-{}[];:'"\,<>./?@#$%^&*_~'''
no_punct=""
for char in l:
  if(char not in punctuations):
    no_punct = no_punct+char
no_punct
```

```python
str1 = no_punct
```

# **Tokenization** 

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
nltk.word_tokenize(str1)
```

```python
str1.split(' ')
```

# **Removing** **stopwords** 

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))
```

```python
print(STOPWORDS)
```

```python
m=""
for i in str1:
  if(i not in STOPWORDS):
    m = m+i
m
```

```python
str2=m
```

## **Removing** **Emojis**

```python
pip install clean-text
```

```python
#import clean function
from cleantext import clean

#provide string with emojis
text = str2

#print text after removing the emojis from it
print(clean(text, no_emoji=True))
```

## **Lemmatization**

```python
text = str2
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

sentence = text
sentence_words = nltk.word_tokenize(sentence)
```

```python
import nltk
nltk.download('omw-1.4')
for word in sentence_words:
  print(wordnet_lemmatizer.lemmatize(word))
```
